**TASK 1: Transform Regression to Binary Classification**

Your mission is to convert the apartment price prediction from a regression problem (predicting exact prices) to a binary classification problem (predicting if an apartment is expensive or not).

**Step-by-step approach:**
1. **Find the median price** - This will serve as your threshold for classification
2. **Create a new binary feature** called `price_class`:
   - For apartments with `last_price` > $113,000 → set `price_class = 1` (expensive)
   - For apartments with `last_price` ≤ $113,000 → set `price_class = 0` (affordable)
3. **Verify your work** by printing the first 5 rows to see the new column

**Why this approach works:**
- Binary classification is often simpler and more interpretable than regression
- Using the median ensures roughly balanced classes
- The $113,000 threshold creates a meaningful distinction between affordable and expensive apartments

</div>

```python
# How to choose what's best types of supervised learning to use here: classification or regression.
# Let's find out the median price
median_price = df['last_price'].median()
print(f"The median price is: {median_price}")
print('Prices above $113,000 are high, while anything below that is considered low.')
```

```python

df.loc[df['last_price'] > 113000, 'price_class'] = 1 
df.loc[df['last_price'] <= 113000, 'price_class'] = 0

print(df.head())
```

**Understanding Decision Trees in Scikit-Learn**

Decision trees are powerful machine learning algorithms that make predictions by learning simple decision rules inferred from the data features. They create a model that predicts the value of a target variable by learning simple decision rules.

**Key Points:**

• **DecisionTreeClassifier**: The sklearn class for binary and multiclass classification problems

• **Tree Structure**: Creates branching paths based on feature values to reach classification decision

• **Interpretability**: One of the most interpretable machine learning algorithms - you can visualize the decision-making process

• **No Data Preprocessing**: Can handle both numerical and categorical features without extensive preprocessing


**Why Decision Trees Work Well for Our Problem:**
- Perfect for binary classification (expensive vs. affordable apartments)
- Can handle mixed data types in our features
- Provides clear rules about what makes an apartment expensive
- Easy to understand and explain the model's decisions

**TASK 3: Training Your Decision Tree Model**

Now that we have our features and target variables prepared, it's time to train our machine learning model. Training is the process where the algorithm learns patterns from your data.

**What happens during training:**
- The DecisionTreeClassifier analyzes the relationship between features (apartment characteristics) and target (price classification)
- It creates decision rules to distinguish between expensive and affordable apartments
- The model learns which features are most important for making predictions

**The fit() method:**
- This is the standard method in scikit-learn for training models
- It takes two arguments: features (X) and target (y)
- After calling fit(), your model will be ready to make predictions

**Ready to train:** Use the `fit()` method on your model object and pass the features and target variables as arguments.

</div>

**TASK 4: Testing Your Model with New Observations**

Now that your Decision Tree model is trained, it's time to test its predictive power! You'll create two new apartment observations with different characteristics and see how well your model classifies them.

**Your Test Cases:**

**Apartment 1 (Luxury):** 12 bedrooms, 900 m² total area, 409.7 m² living area, 112 m² kitchen
- This represents a high-end, spacious apartment

**Apartment 2 (Standard):** 2 bedrooms, 109 m² total area, 32 m² living area, 40.5 m² kitchen  
- This represents a typical family apartment

**Your Mission:**
1. Create feature arrays for both apartments using the same structure as your training data
2. Use the trained model to predict the price class for each apartment
3. Interpret the results: Does the model correctly identify which apartment should be expensive (class 1) vs affordable (class 0)?

**Expected Outcome:**
The model should predict class 1 (expensive) for the luxury apartment and class 0 (affordable) for the standard apartment, demonstrating that it learned meaningful patterns from the training data.

</div>


```python
# Summary of creating a classification model for real estate prices and training it.
# Based on the median price of $113,000, we can classify properties into two categories: high-priced (1) and low-priced (0).
df.loc[df['last_price'] > 113000, 'price_class'] = 1 
df.loc[df['last_price'] <= 113000, 'price_class'] = 0
features = df.drop(['last_price', 'price_class'], axis=1)
target = df['price_class']
model = DecisionTreeClassifier()

model.fit(features, target)
```

```python

new_features = pd.DataFrame(
    [
        [None, None, 2.8, 25, None, 25, 0, 0, 0, None, 0, 30706.0, 7877.0],
        [None, None, 2.75, 25, None, 25, 0, 0, 0, None, 0, 36421.0, 9176.0]
    ],
    columns=features.columns
)
print(new_features.shape)
```

```python
new_features.loc[0, 'total_area'] = 900.0
print(new_features)
print(new_features.shape)
```

```python
# To predict answers, call the predict() method and pass it the table with the features of the new observations.
predictions = model.predict(new_features)
print(predictions)
```

```python
#The first apartment has 12 bedrooms with a total area of 900 m². The living area is 409.7 m², and the kitchen area is 112 m².
new_features.loc[0, 'bedrooms']= 12
new_features.loc[0, 'living_area']= 409.7
new_features.loc[0, 'kitchen_area']= 112.0
print(new_features.shape)
predictions = model.predict(new_features)
print(predictions)  # The prediction changes to 1, indicating that this apartment is classified as high-priced.
```

```python
#The second apartment has 2 bedrooms with a total area of 109 m². The living area is 32 m², and the kitchen area is 40.5 m².
new_features.loc[1,'bedrooms']= 2
new_features.loc[1, 'total_area']= 109.9
new_features.loc[1,'living_area'] = 32
new_features.loc[1,'kitchen_area']= 40.5
predictions = model.predict(new_features)
print(predictions)
```
