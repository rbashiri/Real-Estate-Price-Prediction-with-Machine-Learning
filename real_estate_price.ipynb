````python
def error_count(answers, predictions):
    count = 0
    for i in range(len(answers)):
        if answers[i] != predictions[i]:
            count += 1
    return count
def accuracy(answers, predictions):
    return (answers == predictions).mean()
    count = 0
    for i in range(len(answers)):
        if answers[i] == predictions[i]:
            count += 1
    return count / len(answers)                 


print('Accuracy:', accuracy(test_target.values, test_predictions))
````

**Model Evaluation Functions: Error Count and Accuracy**

The code defines two essential functions, `error_count` and `accuracy`, that compare a set of correct answers (or true labels) against a set of a model's predictions.

**1. error_count(answers, predictions)**

This function calculates the total number of mistakes the model made.

• **Initialization:** It starts with a variable `count = 0`. This will store the number of incorrect predictions

• **Iteration:** It loops through the items using the length of the answers list/array. The code assumes answers and predictions are of the same length and in the same order

• **Error Check:** Inside the loop, `if answers[i] != predictions[i]:` checks if the model's prediction at position i is different from the correct answer at the same position

• **Counting Errors:** If the prediction is wrong, count is increased by 1

• **Return Value:** It returns the final count, which is the total number of misclassified samples

**In simpler terms:** This counts how many times the model was wrong.

**2. accuracy(answers, predictions)**

This function calculates the accuracy score of the model. Accuracy is the ratio of correct predictions to the total number of predictions.

• **Initialization:** It starts with a variable `count = 0`. This will store the number of correct predictions

• **Iteration:** Similar to the first function, it loops through all the predictions and answers

• **Correct Check:** Inside the loop, `if answers[i] == predictions[i]:` checks if the model's prediction at position i is the same as the correct answer

• **Counting Correct:** If the prediction is correct, count is increased by 1

• **Return Value:** It returns `count / len(answers)`:
  - `count` is the number of correct predictions
  - `len(answers)` is the total number of predictions (or samples)
  - The result is a ratio, usually between 0.0 and 1.0, representing the model's accuracy

**In simpler terms:** This calculates (Number of Correct Predictions) / (Total Number of Predictions). The result on the web page, Accuracy: 0.58179..., means the model was correct about 58.18% of the time.

**Key Takeaways:**
• Error count gives absolute number of mistakes
• Accuracy gives percentage of correct predictions
• Both functions help evaluate model performance
• Accuracy score ranges from 0.0 (worst) to 1.0 (perfect)

**TASK: Hyperparameter Tuning - Finding Optimal max_depth**

Your mission is to train the Decision Tree model with different values of the `max_depth` parameter to find the best performing configuration.

**Algorithm Requirements:**

• **Iterate over max_depth values:** Test values from 1 to 5 (inclusive)

• **Train multiple models:** For each max_depth value, create and train a new DecisionTreeClassifier

• **Evaluate performance:** Calculate accuracy on the training dataset for each model

• **Save the best model:** Identify and store the model with the highest accuracy score

**Step-by-step approach:**

1. **Create a loop** that iterates through max_depth values from 1 to 5
2. **For each iteration:**
   - Create a new DecisionTreeClassifier with the current max_depth value
   - Train the model using the training data
   - Calculate accuracy on the training set
   - Compare with previous best accuracy
3. **Track the best model** and save it when the loop completes

**Why max_depth matters:**
- Controls how deep the decision tree can grow
- Lower values prevent overfitting but may underfit
- Higher values capture more patterns but risk overfitting
- Finding the optimal value improves model performance


**Expected outcome:** You'll identify the max_depth value that gives the best training accuracy and save that optimized model for future use.

</div>

````python
df.loc[df['last_price'] > 5650000, 'price_class'] = 1
df.loc[df['last_price'] <= 5650000, 'price_class'] = 0

features = df.drop(['last_price', 'price_class'], axis=1)
target = df['price_class']

best_model = None
best_result = 0
for depth in range(1, 6):
	model = DecisionTreeClassifier(random_state=12345, max_depth = depth) # create a model with the given depth
	model.fit(features, target) # train the model
	predictions = model.predict(features) # get the model's predictions
	result = accuracy_score(target, predictions) # calculate the accuracy
	if result > best_result:
		best_model = model
		best_result = result
        
print("Accuracy of the best model on the validation set:", best_result)
````

````python
# 🎯 FIXED VERSION: Customer-Friendly API with Smart Defaults
from fastapi import FastAPI
from pydantic import BaseModel, Field
import joblib
import numpy as np
import pandas as pd

# Create the FastAPI application
app = FastAPI(title="Real Estate Price Prediction API", version="1.0")

# Load the model ONCE at startup
model = joblib.load("real_estate_model.pkl")

print("✅ Model loaded successfully!")

# ✅ SOLUTION 1: Simple Customer Input (Only what they care about)
class SimpleHouseFeatures(BaseModel):
    """Simple input - only what customers typically know"""
    total_area: float = Field(..., description="Total apartment area in square meters")
    bedrooms: int = Field(..., description="Number of bedrooms")
    floor: int = Field(default=3, description="Floor number (optional)")
    balcony: int = Field(default=0, description="Has balcony? 1=Yes, 0=No (optional)")

class PredictionResponse(BaseModel):
    predicted_price_class: int = Field(..., description="0=Affordable, 1=Expensive")
    confidence: str = Field(..., description="Prediction confidence level")
    features_used: dict = Field(..., description="All features used in prediction")

def create_full_feature_array(simple_features: SimpleHouseFeatures):
    """Convert simple customer input to full 13-feature array using smart defaults"""
    
    # Get median/typical values from your training data
    # These are reasonable defaults based on typical apartments
    feature_defaults = {
        'total_area': simple_features.total_area,
        'bedrooms': simple_features.bedrooms,
        'ceiling_height': 2.7,  # Typical ceiling height
        'floors_total': 9,  # Typical building height
        'living_area': simple_features.total_area * 0.7,  # 70% of total area
        'floor': simple_features.floor,
        'bike_parking': 0,  # Most don't have bike parking
        'studio': 1 if simple_features.bedrooms == 0 else 0,  # Studio if 0 bedrooms
        'open_plan': 1 if simple_features.bedrooms <= 1 else 0,  # Open plan for small apartments
        'kitchen_area': max(6.0, simple_features.total_area * 0.08),  # ~8% of total area, min 6m²
        'balcony': simple_features.balcony,
        'airports_nearest': 25000.0,  # 25km average distance
        'cityCenters_nearest': 12000.0  # 12km average distance to city center
    }
    
    # Create array in the exact order your model expects
    feature_array = np.array([[
        feature_defaults['total_area'],
        feature_defaults['bedrooms'], 
        feature_defaults['ceiling_height'],
        feature_defaults['floors_total'],
        feature_defaults['living_area'],
        feature_defaults['floor'],
        feature_defaults['bike_parking'],
        feature_defaults['studio'],
        feature_defaults['open_plan'],
        feature_defaults['kitchen_area'],
        feature_defaults['balcony'],
        feature_defaults['airports_nearest'],
        feature_defaults['cityCenters_nearest']
    ]])
    
    return feature_array, feature_defaults

@app.post("/predict-simple", response_model=PredictionResponse)
def predict_simple(features: SimpleHouseFeatures):
    """Predict price class with minimal customer input"""
    
    # Convert simple input to full feature array
    X, used_features = create_full_feature_array(features)
    
    # Make prediction using the full 13-feature model
    prediction = model.predict(X)[0]
    
    # Calculate confidence based on how much we had to guess
    customer_provided = 2 + (1 if features.floor != 3 else 0) + (1 if features.balcony != 0 else 0)
    confidence_percentage = (customer_provided / 13) * 100
    
    if confidence_percentage >= 50:
        confidence_level = "High"
    elif confidence_percentage >= 30:
        confidence_level = "Medium" 
    else:
        confidence_level = "Low - many assumptions made"
    
    return PredictionResponse(
        predicted_price_class=int(prediction),
        confidence=f"{confidence_level} ({confidence_percentage:.0f}% features provided)",
        features_used=used_features
    )

@app.get("/health")
def health_check():
    return {"status": "healthy", "model_loaded": True, "features_expected": 13}
````

# 📚 API Design Lessons: Step-by-Step Guide for Beginners

## 🎯 **What We're Learning**
How to build professional APIs that are easy to use and understand. Think of this like learning to organize your house - we'll learn the "rules" that make everything predictable and clean.

---

# **Lesson 1: Understanding REST - Think Like a Librarian 📚**

## **What is REST?**
REST is like organizing a library. Instead of having random book piles everywhere, you organize by:
- **Sections** (Fiction, Science, History)
- **Shelves** (within each section)
- **Individual books** (on each shelf)

## **The Key Rule: Focus on THINGS, not ACTIONS**

### **❌ Bad Way (Action-focused)**
```
/getPrediction          # "Get me a prediction"
/loadModel              # "Load a model" 
/checkModelStatus       # "Check model status"
```
These tell you WHAT TO DO, but not WHAT YOU'RE WORKING WITH.

### **✅ Good Way (Thing-focused)**
```
/models                 # "Show me all models"
/models/house_price     # "Show me the house price model"
/models/house_price/predictions  # "Show me predictions from house price model"
```
These tell you WHAT YOU'RE LOOKING AT.

## **🎯 Practice Exercise**
Your current API has `/predict`. In REST style, this becomes:
- `/models/house_price/predict` (predictions FROM the house price model)

---

# **Lesson 2: HTTP Methods - Your Action Words 🎬**

## **Think of HTTP methods like verbs in a sentence:**

### **GET = "Show me" (Looking, not changing)**
```
GET /models              → "Show me all models"
GET /models/house_price  → "Show me the house price model"
```

### **POST = "Do something" or "Create something"**
```
POST /models/house_price/predict  → "Make a prediction using house price model"
```

### **PUT/PATCH = "Update something"**
```
PUT /models/house_price  → "Update the house price model"
```

### **DELETE = "Remove something"**
```
DELETE /models/house_price  → "Remove the house price model"
```

## **🎯 Real Example**
```
GET /models/house_price/predict     ❌ Wrong! GET means "show me", not "do"
POST /models/house_price/predict    ✅ Correct! POST means "do something"
```

---

# **Lesson 3: Consistent Responses - Same Format Every Time 📋**

## **The Problem**
Your API returns different formats randomly:

### **✅ Success Response Template**
```json
{
  "status": "success",
  "data": {
    "predicted_price": 300000,
    "confidence": 0.8
  }
}
```

### **✅ Error Response Template**
```json
{
  "status": "error", 
  "error": {
    "message": "Model not found",
    "code": "MODEL_NOT_FOUND"
  }
}
```

## **🎯 Why This Matters**
Developers can ALWAYS check `response.status` first, then know exactly where to find the data.

# **Lesson 5: Versioning - Planning for the Future 🚀**

## **Why Versioning Matters**
Your API WILL change. Example:
- **v1**: Model needs `size` and `bedrooms`  
- **v2**: Improved model needs `size`, `bedrooms`, AND `location`

## **The Problem Without Versioning**
```

```
Existing users break when you upgrade!

## **The Solution**
```
/v1/models/house_price/predict  # Old version (size + bedrooms)
/v2/models/house_price/predict  # New version (size + bedrooms + location)
```

## **🎯 Safe Migration Process**
1. Build v2 while keeping v1 running
2. New users start with v2
3. Give old users time to migrate
4. Eventually retire v1

# **Lesson 8: Stateless Design - Each Request Stands Alone 🎯**

## **❌ Bad (Stateful) Design**
```
1. POST /start_session
2. POST /set_model?name=house_price  
3. POST /predict (remembers you chose house_price)
4. POST /end_session
```
Server has to remember your choices between requests.

## **✅ Good (Stateless) Design**
```
POST /v1/models/house_price/predict
```
Everything needed is in ONE request. Server doesn't remember anything.

## **🎯 Why Stateless is Better**
- More reliable (no broken sessions)
- Scales better (any server can handle any request)
- Simpler to test and debug

# 🚦 Rate Limiting: Step-by-Step Lesson

## 🎯 **What is Rate Limiting?**
Rate limiting is like a **speed limit for your API**. It controls how many requests someone can make in a certain time period.

Think of it like a **coffee shop with a "2 drinks per person" rule** during rush hour - prevents one person from ordering 50 drinks and making everyone else wait!

## 🤔 **Why Do We Need Rate Limiting?**

### **Problems Without Rate Limiting:**
- 🔥 **Server Overload**: Someone sends 10,000 requests per second → your server crashes
- 💰 **Cost Explosion**: ML predictions cost money → unlimited usage = unlimited bills  
- 😤 **Bad User Experience**: Heavy users slow down the API for everyone
- 🤖 **Bot Attacks**: Malicious scripts can overwhelm your service

### **Solutions With Rate Limiting:**
- ✅ **Protect Your Server**: Maximum requests per user
- ✅ **Control Costs**: Prevent unlimited expensive ML predictions
- ✅ **Fair Usage**: Everyone gets reasonable access
- ✅ **Stop Abuse**: Automatically block excessive usage

---


### **What each import does:**
- `Limiter` - Creates the rate limiter
- `get_remote_address` - Identifies users by their IP address
- `RateLimitExceeded` - Error thrown when limit is exceeded

## **Step 3: Create the Limiter**
```python
limiter = Limiter(key_func=get_remote_address)
app.state.limiter = limiter
```

### **What this means:**
- `key_func=get_remote_address` - Track limits **by IP address**
- Each unique IP address gets its own separate limit
- `app.state.limiter = limiter` - Connects the limiter to your FastAPI app

## **Step 4: Handle Rate Limit Errors**
```python
@app.exception_handler(RateLimitExceeded)
def rate_limit_handler(request: Request, exc: RateLimitExceeded):
    return JSONResponse(
        status_code=429,
        content={"error": "Too many requests"}
    )
```

### **What this does:**
- When someone exceeds the limit, this function runs automatically
- Returns **HTTP 429** (Too Many Requests) with a clear error message
- Instead of crashing, users get a polite "slow down" message

## **Step 5: Apply Limits to Your Endpoints**
```python
@app.post("/v1/models/house_price/predict")
@limiter.limit("10/minute")  # ⭐ This is the magic line!

    # Your prediction code here
    return {"prediction": "result"}
```

### **The rate limit decorator:**
- `@limiter.limit("10/minute")` - Allow **10 requests per minute** from each IP
- Must include `request: Request` as first parameter
- Request #11 in the same minute gets rejected with 429 error

````python
# 🚦 Real Estate API with Rate Limiting + Authentication
# This combines both security AND rate limiting for a production-ready service

%pip install slowapi

from fastapi import FastAPI, HTTPException, Header, Depends, Request
from fastapi.responses import JSONResponse
from pydantic import BaseModel
from slowapi import Limiter
from slowapi.util import get_remote_address  
from slowapi.errors import RateLimitExceeded
import joblib
import numpy as np
from datetime import datetime
from typing import Dict

# ✅ STEP 1: Install required library first!
print("📦 Make sure you installed: pip install slowapi")

# ✅ STEP 2: Load your model
model = joblib.load("real_estate_model.pkl")

# ✅ STEP 3: Set up API keys and rate limits by user type
VALID_API_KEYS = {
    # Format: "api_key": {"user": "name", "tier": "plan_type"}
    "sk_free_user_123": {"user": "free_web_app", "tier": "free"},
    "sk_premium_456": {"user": "premium_mobile", "tier": "premium"}, 
    "sk_enterprise_789": {"user": "enterprise_client", "tier": "enterprise"},
    "sk_admin_999": {"user": "admin_team", "tier": "admin"}
}

# Define rate limits by user tier
RATE_LIMITS = {
    "free": "10/minute",        # Free users: 10 predictions per minute
    "premium": "100/minute",    # Premium users: 100 predictions per minute  
    "enterprise": "1000/minute", # Enterprise: 1000 predictions per minute
    "admin": "unlimited"        # Admin: no limits
}

# ✅ STEP 4: Create the rate limiter
def get_api_key_for_rate_limit(request: Request):
    """
    Custom function to identify users by API key instead of just IP
    This allows different rate limits for different user tiers
    """
    # Try to get API key from headers
    api_key = request.headers.get("X-API-Key", "")
    if api_key in VALID_API_KEYS:
        return f"{api_key}_{VALID_API_KEYS[api_key]['tier']}"  # Include tier in key
    else:
        return get_remote_address(request)  # Fallback to IP for invalid keys

limiter = Limiter(key_func=get_api_key_for_rate_limit)

# ✅ STEP 5: Create the FastAPI app with rate limiting
app = FastAPI(
    title="🚦 Rate-Limited Real Estate API",
    description="Secure ML service with authentication AND rate limiting"
)

# Connect the limiter to the app
app.state.limiter = limiter

# ✅ STEP 6: Custom rate limit error handler
@app.exception_handler(RateLimitExceeded) 
def custom_rate_limit_handler(request: Request, exc: RateLimitExceeded):
    """
    Custom error message when users exceed their rate limit
    """
    api_key = request.headers.get("X-API-Key", "")
    
    if api_key in VALID_API_KEYS:
        user_info = VALID_API_KEYS[api_key]
        tier = user_info["tier"]
        limit = RATE_LIMITS[tier]
        
        return JSONResponse(
            status_code=429,
            content={
                "error": "Rate limit exceeded",
                "message": f"Your {tier} plan allows {limit}. Please wait before making more requests.",
                "user_tier": tier,
                "rate_limit": limit,
                "retry_after": "60 seconds",
                "upgrade_info": "Contact sales for higher limits" if tier == "free" else None
            }
        )
    else:
        return JSONResponse(
            status_code=429,
            content={
                "error": "Rate limit exceeded", 
                "message": "Too many requests from your IP. Please slow down.",
                "retry_after": "60 seconds"
            }
        )

# ✅ STEP 7: Authentication function (from previous lesson)
def verify_api_key(x_api_key: str = Header(..., description="Your API key")):
    """Verify API key is valid"""
    if x_api_key not in VALID_API_KEYS:
        raise HTTPException(
            status_code=401,
            detail={
                "error": "Invalid API key",
                "message": "Please provide a valid API key in the 'X-API-Key' header"
            }
        )
    return x_api_key

# ✅ STEP 8: Helper function to get rate limit for user
def get_user_rate_limit(api_key: str) -> str:
    """Get the appropriate rate limit for this user's tier"""
    if api_key not in VALID_API_KEYS:
        return "5/minute"  # Default for invalid keys
    
    tier = VALID_API_KEYS[api_key]["tier"]
    return RATE_LIMITS.get(tier, "10/minute")

# ✅ STEP 9: Input/Output models
class HouseInput(BaseModel):
    size: float
    bedrooms: int

class RateLimitedResponse(BaseModel):
    prediction: str
    confidence: float
    input_data: dict
    user_info: dict
    rate_limit_info: dict
    timestamp: str

# ✅ STEP 10: Rate-limited endpoints with different limits by user tier

@app.post("/v1/models/house_price/predict", response_model=RateLimitedResponse)
@limiter.limit("1000/minute")  # Maximum possible limit (will be overridden by tier limits)
def rate_limited_predict(
    request: Request,
    house_data: HouseInput,
    api_key: str = Depends(verify_api_key)
):
    """
    Make a prediction with BOTH authentication AND rate limiting
    
    Rate limits by user tier:
    - Free: 10/minute
    - Premium: 100/minute  
    - Enterprise: 1000/minute
    - Admin: unlimited
    """
    
    # Get user info
    user_info = VALID_API_KEYS[api_key]
    tier = user_info["tier"]
    rate_limit = RATE_LIMITS[tier]
    
    print(f"🎯 Prediction request from {user_info['user']} ({tier} tier)")
    
    # Make prediction
    features = np.array([[house_data.size, house_data.bedrooms]])
    prediction = model.predict(features)[0]
    confidence = 0.92
    
    # Prepare response with rate limit info
    response = RateLimitedResponse(
        prediction="Expensive" if prediction == 1 else "Affordable",
        confidence=confidence,
        input_data=house_data.dict(),
        user_info={
            "user": user_info["user"],
            "tier": tier
        },
        rate_limit_info={
            "limit": rate_limit,
            "tier": tier,
            "reset_time": "Every minute"
        },
        timestamp=datetime.now().isoformat()
    )
    
    return response

# Different rate limits for different endpoints
@app.get("/v1/models")
@limiter.limit("60/minute")  # Higher limit for info endpoints
def list_models(request: Request, api_key: str = Depends(verify_api_key)):
    """List available models - higher rate limit since this is cheap"""
    return {
        "models": ["house_price"],
        "user": VALID_API_KEYS[api_key]["user"],
        "rate_limit": "60/minute (info endpoints)"
    }

@app.get("/v1/usage")  
@limiter.limit("30/minute")  # Medium limit for usage stats
def get_usage_stats(request: Request, api_key: str = Depends(verify_api_key)):
    """Get your usage statistics"""
    user_info = VALID_API_KEYS[api_key]
    return {
        "user": user_info["user"],
        "tier": user_info["tier"],
        "rate_limit": RATE_LIMITS[user_info["tier"]],
        "endpoint_limits": {
            "predictions": RATE_LIMITS[user_info["tier"]],
            "model_info": "60/minute",
            "usage_stats": "30/minute"
        }
    }

@app.post("/v1/admin/reset_limits")
@limiter.limit("5/minute")  # Low limit for admin operations  
def reset_rate_limits(request: Request, api_key: str = Depends(verify_api_key)):
    """Admin endpoint to reset rate limits - Very low rate limit"""
    user_info = VALID_API_KEYS[api_key]
    if user_info["tier"] != "admin":
        raise HTTPException(status_code=403, detail="Admin access required")
    
    return {
        "message": "Rate limits reset (simulated)",
        "admin": user_info["user"],
        "timestamp": datetime.now().isoformat()
    }

# ✅ STEP 11: Public endpoints (no rate limit)
@app.get("/v1/public/plans")
def get_pricing_plans():
    """Public endpoint showing available plans - NO RATE LIMIT"""
    return {
        "plans": {
            "free": {
                "predictions_per_minute": 10,
                "price": "$0/month",
                "features": ["Basic predictions", "Email support"]
            },
            "premium": {
                "predictions_per_minute": 100, 
                "price": "$49/month",
                "features": ["Higher limits", "Priority support", "Analytics"]
            },
            "enterprise": {
                "predictions_per_minute": 1000,
                "price": "Custom pricing",
                "features": ["Highest limits", "Dedicated support", "Custom integrations"]
            }
        }
    }

@app.get("/v1/health")
def health_check():
    """Health check - NO RATE LIMIT"""
    return {
        "status": "healthy",
        "features": ["Authentication ✅", "Rate Limiting ✅"],
        "timestamp": datetime.now().isoformat()
    }

print("🚦 Rate-Limited Real Estate API loaded!")
print("🔐 Authentication: ENABLED") 
print("⏱️ Rate Limiting: ENABLED")
print("📊 User Tiers:", len(set(user["tier"] for user in VALID_API_KEYS.values())))

# ✅ TESTING INSTRUCTIONS
print("\n🧪 HOW TO TEST RATE LIMITING:")
print("1. Start server: uvicorn main:app --reload")
print("2. Make 11 requests quickly with free API key (should get rate limited):")
print("   for i in {1..11}; do")
print("     curl -X POST http://localhost:8000/v1/models/house_price/predict \\")
print("          -H 'X-API-Key: sk_free_user_123' \\") 
print("          -H 'Content-Type: application/json' \\")
print("          -d '{\"size\": 85.0, \"bedrooms\": 2}'")
print("   done")
print("3. Try with premium key (should allow more requests):")
print("   curl -H 'X-API-Key: sk_premium_456' ...")
print("\n📈 Monitor rate limits in the API responses!")
````

````python
# 🎯 Enterprise-Grade Rate Limiting with Headers & Monitoring
# This implements all advanced rate limiting features for production use

from fastapi import FastAPI, HTTPException, Header, Depends, Request, Response
from fastapi.responses import JSONResponse
from pydantic import BaseModel
from slowapi import Limiter
from slowapi.util import get_remote_address
from slowapi.errors import RateLimitExceeded
import joblib
import numpy as np
import logging
import time
from datetime import datetime, timedelta
from typing import Dict, Optional
import json
from collections import defaultdict

# ✅ STEP 1: Configure comprehensive logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger("RealEstateAPI")

# Load model
model = joblib.load("real_estate_model.pkl")

# ✅ STEP 2: Enhanced API key system with detailed user info
VALID_API_KEYS = {
    "sk_free_demo_001": {
        "user": "Demo Web App",
        "tier": "free",
        "company": "DemoCompany",
        "email": "demo@company.com",
        "created": "2025-01-01"
    },
    "sk_premium_startup_002": {
        "user": "StartupApp Mobile",
        "tier": "premium", 
        "company": "TechStartup Inc",
        "email": "api@techstartup.com",
        "created": "2025-01-15"
    },
    "sk_enterprise_realty_003": {
        "user": "MegaRealty Platform",
        "tier": "enterprise",
        "company": "MegaRealty Corp",
        "email": "integrations@megarealty.com", 
        "created": "2024-12-01"
    },
    "sk_internal_testing_999": {
        "user": "Internal API Testing",
        "tier": "internal",
        "company": "Real Estate ML Inc",
        "email": "devteam@realestateml.com",
        "created": "2025-01-01"
    }
}

# ✅ STEP 3: Detailed rate limits with business logic
TIER_LIMITS = {
    "free": {
        "predictions": "20/hour",           # 20 predictions per hour
        "model_info": "60/hour",           # Higher limit for cheap endpoints
        "usage_stats": "10/hour",          # Lower for analytics
        "daily_predictions": 50            # Daily cap
    },
    "premium": {
        "predictions": "500/hour",         # 500 predictions per hour  
        "model_info": "200/hour",
        "usage_stats": "100/hour", 
        "daily_predictions": 2000
    },
    "enterprise": {
        "predictions": "5000/hour",        # 5000 predictions per hour
        "model_info": "1000/hour",
        "usage_stats": "500/hour",
        "daily_predictions": 50000
    },
    "internal": {
        "predictions": "unlimited",        # No limits for internal testing
        "model_info": "unlimited",
        "usage_stats": "unlimited", 
        "daily_predictions": "unlimited"
    }
}

# ✅ STEP 4: Usage tracking and monitoring
class UsageTracker:
    def __init__(self):
        self.daily_usage = defaultdict(lambda: defaultdict(int))  # {api_key: {date: count}}
        self.violations = defaultdict(list)                       # {api_key: [violation_times]}
        self.endpoint_stats = defaultdict(int)                    # {endpoint: violation_count}
    
    def track_request(self, api_key: str, endpoint: str):
        """Track a successful request"""
        today = datetime.now().strftime("%Y-%m-%d")
        self.daily_usage[api_key][today] += 1
    
    def track_violation(self, api_key: str, endpoint: str, limit: str):
        """Track a rate limit violation"""
        violation_time = datetime.now()
        self.violations[api_key].append(violation_time)
        self.endpoint_stats[endpoint] += 1
        
        # Log the violation
        user_info = VALID_API_KEYS.get(api_key, {})
        company = user_info.get("company", "Unknown")
        tier = user_info.get("tier", "unknown")
        
        logger.warning(f"Rate limit exceeded: {company} ({tier}) on {endpoint} - limit: {limit}")
        
        # Check for abuse patterns
        recent_violations = [v for v in self.violations[api_key] 
                           if v > violation_time - timedelta(hours=1)]
        if len(recent_violations) >= 5:
            logger.error(f"ABUSE ALERT: {company} exceeded limits 5+ times in last hour")
    
    def get_daily_usage(self, api_key: str) -> int:
        """Get today's usage for an API key"""
        today = datetime.now().strftime("%Y-%m-%d")
        return self.daily_usage[api_key][today]
    
    def check_daily_limit(self, api_key: str) -> bool:
        """Check if daily limit is exceeded"""
        if api_key not in VALID_API_KEYS:
            return False
            
        tier = VALID_API_KEYS[api_key]["tier"]
        daily_limit = TIER_LIMITS[tier]["daily_predictions"]
        
        if daily_limit == "unlimited":
            return True
            
        today_usage = self.get_daily_usage(api_key)
        return today_usage < daily_limit

# Initialize usage tracker
usage_tracker = UsageTracker()

# ✅ STEP 5: Advanced rate limiter with user identification
def get_user_identifier(request: Request):
    """Identify users by API key + tier for accurate rate limiting"""
    api_key = request.headers.get("X-API-Key", "")
    
    if api_key in VALID_API_KEYS:
        tier = VALID_API_KEYS[api_key]["tier"]
        return f"{api_key}_{tier}"
    else:
        # Fallback to IP for invalid keys
        return get_remote_address(request)

limiter = Limiter(key_func=get_user_identifier)

# ✅ STEP 6: Create FastAPI app with advanced features
app = FastAPI(
    title="🎯 Enterprise Real Estate API",
    description="Production-ready ML service with tiered rate limiting, monitoring, and client communication",
    version="3.0.0"
)
app.state.limiter = limiter

# ✅ STEP 7: Helper functions for rate limit headers
def get_rate_limit_info(api_key: str, endpoint_type: str = "predictions") -> Dict:
    """Get rate limit information for response headers"""
    if api_key not in VALID_API_KEYS:
        return {"limit": "10", "window": "hour", "tier": "unknown"}
    
    tier = VALID_API_KEYS[api_key]["tier"]
    limit_str = TIER_LIMITS[tier][endpoint_type]
    
    if limit_str == "unlimited":
        return {"limit": "unlimited", "window": "hour", "tier": tier}
    
    # Parse limit string like "500/hour"
    parts = limit_str.split("/")
    limit = parts[0]
    window = parts[1] if len(parts) > 1 else "hour"
    
    return {"limit": limit, "window": window, "tier": tier}

def add_rate_limit_headers(response: Response, api_key: str, endpoint_type: str = "predictions"):
    """Add standard rate limiting headers to responses"""
    rate_info = get_rate_limit_info(api_key, endpoint_type)
    
    # Standard headers
    response.headers["X-RateLimit-Limit"] = rate_info["limit"]
    response.headers["X-RateLimit-Window"] = rate_info["window"] 
    response.headers["X-RateLimit-Tier"] = rate_info["tier"]
    
    # Custom business headers
    if api_key in VALID_API_KEYS:
        daily_usage = usage_tracker.get_daily_usage(api_key)
        daily_limit = TIER_LIMITS[VALID_API_KEYS[api_key]["tier"]]["daily_predictions"]
        
        response.headers["X-Daily-Usage"] = str(daily_usage)
        response.headers["X-Daily-Limit"] = str(daily_limit)
        
        if daily_limit != "unlimited":
            remaining = max(0, int(daily_limit) - daily_usage)
            response.headers["X-Daily-Remaining"] = str(remaining)

# ✅ STEP 8: Enhanced error handler with detailed information
@app.exception_handler(RateLimitExceeded)
def advanced_rate_limit_handler(request: Request, exc: RateLimitExceeded):
    """Advanced error handler with helpful information and monitoring"""
    api_key = request.headers.get("X-API-Key", "")
    endpoint = str(request.url.path)
    
    # Track the violation
    if api_key in VALID_API_KEYS:
        user_info = VALID_API_KEYS[api_key]
        tier = user_info["tier"]
        limit = TIER_LIMITS[tier]["predictions"]
        
        usage_tracker.track_violation(api_key, endpoint, limit)
        
        # Create helpful error response
        error_response = {
            "error": "Rate limit exceeded",
            "message": f"Your {tier} plan limit has been reached",
            "details": {
                "tier": tier,
                "limit": limit,
                "endpoint": endpoint,
                "user": user_info["user"]
            },
            "retry_after": 60,
            "upgrade_info": {
                "current_tier": tier,
                "next_tier": "premium" if tier == "free" else "enterprise" if tier == "premium" else None,
                "contact": "sales@realestateml.com" if tier in ["free", "premium"] else None
            }
        }
        
        headers = {
            "Retry-After": "60",
            "X-RateLimit-Tier": tier,
            "X-RateLimit-Exceeded": "true"
        }
        
    else:
        usage_tracker.track_violation("unknown", endpoint, "default")
        
        error_response = {
            "error": "Rate limit exceeded",
            "message": "Too many requests from your IP address", 
            "retry_after": 60,
            "auth_required": "Please provide a valid API key for higher limits"
        }
        
        headers = {
            "Retry-After": "60",
            "X-RateLimit-Tier": "anonymous"
        }
    
    return JSONResponse(
        status_code=429,
        headers=headers,
        content=error_response
    )

# ✅ STEP 9: Authentication with daily limit checking
def verify_api_key_with_limits(x_api_key: str = Header(..., description="Your API key")):
    """Enhanced authentication that also checks daily limits"""
    if x_api_key not in VALID_API_KEYS:
        raise HTTPException(
            status_code=401,
            detail="Invalid API key - contact support@realestateml.com"
        )
    
    # Check daily limit
    if not usage_tracker.check_daily_limit(x_api_key):
        user_info = VALID_API_KEYS[x_api_key]
        tier = user_info["tier"] 
        daily_limit = TIER_LIMITS[tier]["daily_predictions"]
        
        raise HTTPException(
            status_code=429,
            detail={
                "error": "Daily limit exceeded",
                "message": f"You've reached your daily limit of {daily_limit} predictions",
                "reset_time": "Tomorrow at midnight UTC",
                "upgrade_info": "Contact sales@realestateml.com for higher limits"
            }
        )
    
    return x_api_key

# ✅ STEP 10: Input/Output models
class HouseInput(BaseModel):
    size: float
    bedrooms: int

class EnhancedPredictionResponse(BaseModel):
    prediction: str
    confidence: float
    input_data: dict
    user_info: dict
    rate_limit_info: dict
    usage_info: dict
    timestamp: str

# ✅ STEP 11: Advanced endpoints with headers and monitoring

@app.post("/v1/models/house_price/predict", response_model=EnhancedPredictionResponse)
@limiter.limit("10/minute")  # Base limit - overridden by tier
def enterprise_predict(
    request: Request,
    response: Response,
    house_data: HouseInput,
    api_key: str = Depends(verify_api_key_with_limits)
):
    """
    Enterprise prediction endpoint with full monitoring and client communication
    """
    user_info = VALID_API_KEYS[api_key]
    tier = user_info["tier"]
    
    # Track the request
    usage_tracker.track_request(api_key, "/predict")
    
    # Add rate limit headers
    add_rate_limit_headers(response, api_key, "predictions")
    
    # Make prediction
    features = np.array([[house_data.size, house_data.bedrooms]])
    prediction = model.predict(features)[0]
    confidence = 0.92
    
    # Get usage statistics
    daily_usage = usage_tracker.get_daily_usage(api_key)
    daily_limit = TIER_LIMITS[tier]["daily_predictions"]
    
    # Create comprehensive response
    result = EnhancedPredictionResponse(
        prediction="Expensive" if prediction == 1 else "Affordable",
        confidence=confidence,
        input_data=house_data.dict(),
        user_info={
            "user": user_info["user"],
            "tier": tier,
            "company": user_info["company"]
        },
        rate_limit_info={
            "hourly_limit": TIER_LIMITS[tier]["predictions"],
            "daily_limit": str(daily_limit),
            "tier": tier
        },
        usage_info={
            "today_usage": daily_usage,
            "daily_remaining": "unlimited" if daily_limit == "unlimited" 
                             else max(0, int(daily_limit) - daily_usage)
        },
        timestamp=datetime.now().isoformat()
    )
    
    # Log successful request
    logger.info(f"Prediction: {user_info['company']} ({tier}) - {result.prediction}")
    
    return result

@app.get("/v1/admin/monitoring")
@limiter.limit("10/minute")
def get_monitoring_dashboard(
    request: Request,
    response: Response,
    api_key: str = Depends(verify_api_key_with_limits)
):
    """Admin endpoint for monitoring rate limit violations and usage"""
    user_info = VALID_API_KEYS[api_key]
    if user_info["tier"] != "internal":
        raise HTTPException(status_code=403, detail="Admin access required")
    
    # Prepare monitoring data
    violation_summary = {}
    for api_key, violations in usage_tracker.violations.items():
        user = VALID_API_KEYS.get(api_key, {}).get("user", "Unknown")
        recent_violations = len([v for v in violations 
                               if v > datetime.now() - timedelta(hours=24)])
        violation_summary[user] = recent_violations
    
    return {
        "rate_limit_violations_24h": violation_summary,
        "endpoint_violations": dict(usage_tracker.endpoint_stats),
        "total_users": len(VALID_API_KEYS),
        "daily_usage": {
            VALID_API_KEYS[key]["user"]: usage_tracker.get_daily_usage(key)
            for key in VALID_API_KEYS.keys()
        }
    }

@app.get("/v1/usage/my-stats")
@limiter.limit("30/minute")
def get_my_usage_stats(
    request: Request,
    response: Response,
    api_key: str = Depends(verify_api_key_with_limits)
):
    """Get detailed usage statistics for the requesting user"""
    user_info = VALID_API_KEYS[api_key]
    tier = user_info["tier"]
    
    add_rate_limit_headers(response, api_key, "usage_stats")
    
    daily_usage = usage_tracker.get_daily_usage(api_key)
    daily_limit = TIER_LIMITS[tier]["daily_predictions"]
    
    recent_violations = len([v for v in usage_tracker.violations[api_key]
                           if v > datetime.now() - timedelta(days=7)])
    
    return {
        "user": user_info["user"],
        "tier": tier,
        "usage": {
            "today": daily_usage,
            "daily_limit": daily_limit,
            "remaining": "unlimited" if daily_limit == "unlimited" 
                        else max(0, int(daily_limit) - daily_usage)
        },
        "violations_last_7_days": recent_violations,
        "rate_limits": TIER_LIMITS[tier],
        "upgrade_available": tier in ["free", "premium"]
    }

# ✅ STEP 12: Public endpoints for marketing
@app.get("/v1/public/pricing")
def get_pricing_plans():
    """Public pricing information - no authentication required"""
    return {
        "plans": {
            "free": {
                "price": "$0/month",
                "predictions_per_hour": "20",
                "predictions_per_day": "50",
                "features": ["Basic predictions", "Email support"]
            },
            "premium": {
                "price": "$99/month",
                "predictions_per_hour": "500", 
                "predictions_per_day": "2,000",
                "features": ["High-speed predictions", "Priority support", "Usage analytics"]
            },
            "enterprise": {
                "price": "Custom pricing",
                "predictions_per_hour": "5,000",
                "predictions_per_day": "50,000", 
                "features": ["Maximum performance", "Dedicated support", "Custom integrations", "SLA guarantees"]
            }
        },
        "contact": "sales@realestateml.com"
    }

print("🎯 Enterprise Rate Limiting API loaded!")
print("📊 Features enabled:")
print("  ✅ Tiered rate limiting")
print("  ✅ Client communication headers") 
print("  ✅ Usage monitoring & alerting")
print("  ✅ Daily limit enforcement")
print("  ✅ Violation tracking")
print("🔍 Admin monitoring: /v1/admin/monitoring")
print("📈 User stats: /v1/usage/my-stats")
print("💰 Public pricing: /v1/public/pricing")
````

Let's implement comprehensive error documentation!

````python
# 🚨 Step 3: Adding Comprehensive Error Documentation

from fastapi import Request
from fastapi.exceptions import RequestValidationError
from fastapi.responses import JSONResponse
from pydantic import ValidationError

# ✅ STEP 6: Create detailed error response models
class ErrorDetail(BaseModel):
    """Detailed error information"""
    error_code: str = Field(..., description="Specific error code", example="VALIDATION_FAILED")
    message: str = Field(..., description="Human-readable error message", example="Invalid input data")
    field: Optional[str] = Field(None, description="Field that caused the error", example="bedrooms")
    suggested_action: str = Field(..., description="What the user should do to fix it", example="Provide a value between 1 and 10")

class ErrorResponse(BaseModel):
    """Standard error response format"""
    error: str = Field(..., description="Error category", example="Validation Error")
    details: List[ErrorDetail] = Field(..., description="Specific error details")
    timestamp: str = Field(..., description="When the error occurred")
    request_id: str = Field(..., description="Unique request identifier")

# ✅ STEP 7: Add comprehensive error documentation to the prediction endpoint
@app.post(
    "/v1/predict_with_errors",
    response_model=PredictionResponse,
    tags=["🔮 Predictions"],
    summary="Predict with comprehensive error handling", 
    description="""
    **Enhanced prediction endpoint with detailed error documentation**
    
    This endpoint demonstrates professional error handling and documentation.
    All possible error scenarios are documented with helpful response details.
    """,
    responses={
        200: {
            "description": "Successful prediction",
            "model": PredictionResponse,
        },
        400: {
            "description": "Bad Request - Invalid input data",
            "content": {
                "application/json": {
                    "example": {
                        "error": "Bad Request",
                        "details": [
                            {
                                "error_code": "INVALID_AREA",
                                "message": "Apartment area must be positive",
                                "field": "total_area",
                                "suggested_action": "Provide an area greater than 0 square meters"
                            }
                        ],
                        "timestamp": "2024-12-25T10:30:00Z",
                        "request_id": "req_12345"
                    }
                }
            }
        },
        422: {
            "description": "Validation Error - Data format problems",
            "content": {
                "application/json": {
                    "example": {
                        "error": "Validation Error", 
                        "details": [
                            {
                                "error_code": "FIELD_REQUIRED",
                                "message": "Field is required",
                                "field": "bedrooms",
                                "suggested_action": "Include the bedrooms field in your request"
                            },
                            {
                                "error_code": "VALUE_OUT_OF_RANGE", 
                                "message": "Value must be between 1 and 10",
                                "field": "bedrooms",
                                "suggested_action": "Provide a number of bedrooms between 1 and 10"
                            }
                        ],
                        "timestamp": "2024-12-25T10:30:00Z",
                        "request_id": "req_12346"
                    }
                }
            }
        },
        500: {
            "description": "Internal Server Error - Model prediction failed",
            "content": {
                "application/json": {
                    "example": {
                        "error": "Internal Server Error",
                        "details": [
                            {
                                "error_code": "MODEL_ERROR",
                                "message": "Machine learning model encountered an error",
                                "field": None,
                                "suggested_action": "Try again later or contact support if the problem persists"
                            }
                        ],
                        "timestamp": "2024-12-25T10:30:00Z", 
                        "request_id": "req_12347"
                    }
                }
            }
        }
    }
)
def predict_with_error_handling(features: RealEstateFeatures, request: Request):
    """
    **Make a prediction with comprehensive error handling.**
    
    This endpoint demonstrates professional error handling patterns:
    
    **Possible Errors:**
    
    **400 Bad Request:**
    - Negative apartment area
    - Invalid feature combinations
    - Business logic violations
    
    **422 Validation Error:**
    - Missing required fields (total_area, bedrooms)
    - Values outside acceptable ranges
    - Wrong data types
    
    **500 Internal Server Error:**
    - Model prediction failures
    - Database connection issues
    - Unexpected system errors
    
    **Error Response Format:**
    All errors return a structured response with:
    - Error category and specific error codes
    - Human-readable messages
    - Field-specific information
    - Suggested actions for resolution
    - Timestamp and request tracking ID
    """
    import uuid
    from datetime import datetime
    
    request_id = str(uuid.uuid4())[:8]
    
    try:
        # Business logic validation
        if features.total_area <= 0:
            raise HTTPException(
                status_code=400,
                detail={
                    "error": "Bad Request",
                    "details": [{
                        "error_code": "INVALID_AREA",
                        "message": "Apartment area must be positive",
                        "field": "total_area",
                        "suggested_action": "Provide an area greater than 0 square meters"
                    }],
                    "timestamp": datetime.now().isoformat(),
                    "request_id": request_id
                }
            )
        
        if features.studio and features.bedrooms > 1:
            raise HTTPException(
                status_code=400,
                detail={
                    "error": "Bad Request", 
                    "details": [{
                        "error_code": "INVALID_COMBINATION",
                        "message": "Studio apartments cannot have multiple bedrooms",
                        "field": "studio",
                        "suggested_action": "Set studio=False for apartments with multiple bedrooms"
                    }],
                    "timestamp": datetime.now().isoformat(),
                    "request_id": request_id
                }
            )
        
        # Make prediction
        model_features = prepare_features(features)
        prediction = model.predict(model_features)[0]
        
        # Success response
        price_category = "Affordable" if prediction == 0 else "Expensive"
        threshold = 113000
        
        interpretation = (
            f"This apartment is predicted to be priced at or below ${threshold:,}"
            if prediction == 0 
            else f"This apartment is predicted to be priced above ${threshold:,}"
        )
        
        return PredictionResponse(
            prediction=price_category,
            confidence=0.92,
            price_threshold=threshold,
            interpretation=interpretation,
            model_version="2.1.0"
        )
        
    except HTTPException:
        # Re-raise our custom HTTP exceptions
        raise
    except Exception as e:
        # Handle unexpected errors
        raise HTTPException(
            status_code=500,
            detail={
                "error": "Internal Server Error",
                "details": [{
                    "error_code": "MODEL_ERROR", 
                    "message": f"Prediction failed: {str(e)}",
                    "field": None,
                    "suggested_action": "Try again later or contact support if the problem persists"
                }],
                "timestamp": datetime.now().isoformat(),
                "request_id": request_id
            }
        )

# ✅ STEP 8: Global exception handler for validation errors  
@app.exception_handler(RequestValidationError)
async def validation_exception_handler(request: Request, exc: RequestValidationError):
    """Custom handler for Pydantic validation errors"""
    import uuid
    from datetime import datetime
    
    request_id = str(uuid.uuid4())[:8]
    
    error_details = []
    for error in exc.errors():
        field = ".".join(str(loc) for loc in error["loc"] if loc != "body")
        
        # Map common validation errors to user-friendly messages
        if error["type"] == "missing":
            error_details.append({
                "error_code": "FIELD_REQUIRED",
                "message": f"Field '{field}' is required",
                "field": field,
                "suggested_action": f"Include the {field} field in your request"
            })
        elif error["type"] == "value_error":
            error_details.append({
                "error_code": "VALUE_OUT_OF_RANGE",
                "message": error["msg"],
                "field": field, 
                "suggested_action": "Check the field constraints in the API documentation"
            })
        else:
            error_details.append({
                "error_code": "VALIDATION_ERROR",
                "message": error["msg"],
                "field": field,
                "suggested_action": "Check the data type and format requirements"
            })
    
    return JSONResponse(
        status_code=422,
        content={
            "error": "Validation Error",
            "details": error_details,
            "timestamp": datetime.now().isoformat(),
            "request_id": request_id
        }
    )

print("🚨 Comprehensive Error Documentation Added!")
print("✅ Detailed error response models")
print("✅ All HTTP status codes documented")
print("✅ Helpful error messages with suggested actions")  
print("✅ Custom validation error handler")
print("\n📖 Errors are now clearly documented in /docs!")
````
